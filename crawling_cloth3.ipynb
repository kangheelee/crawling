```python
user_id = []
for n in range(0,23):
    base_url = f'https://cafe.naver.com/soho/ArticleList.nhn?search.clubid=10094408&search.menuid=648&userDisplay=50&search.boardtype=L&search.specialmenutype=&search.questionTab=A&search.totalCount=501&search.page={n}'
    driver.get(base_url)
    #iframe으로 프레임 전환
    driver.switch_to.frame('cafe_main')
    user_id.extend([ i.get_attribute('onclick') for i in driver.find_elements_by_css_selector('td:nth-child(3) > div > table > tbody > tr > td > a')])
```

```python
id_list = []
for user in user_id:
        user1 = user.split(',')
        user2 = user1[1]
        user3 = user2[2:-1]+ str('@naver.com')
        nickname = user1[3]
        nickname = nickname[1:-1]
        id_list.append({'user': user3, 'nickname': nickname})
```
