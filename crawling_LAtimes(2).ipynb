```python
# _*_ coding: UTF-8 _*_
#coding:utf-8
from urllib.request import urlopen
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd
import csv
from selenium.webdriver.remote.webelement import WebElement
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import re
import numpy as np
import openpyxl
import csv
```

```python
df = pd.read_excel('latimes_href2.xlsx') # df is pandas.DataFrame
```

```python
url_df = df["link"]
list_dir = []
for url in url_df:
    list_dir.append(url)
list_dir
```

```python
list_dir6 = list_dir[0:10000]
list_dir7 = list_dir[10000:20000]
list_dir8 = list_dir[20000:30000]
list_dir9 = list_dir[30000:40000]
list_dir10 = list_dir[40000:50000]
```

```python
n=0
res_list = []
for href in list_dir7:
    n = n + 1
    content_total = []
    try:
        html = urlopen(href)
        soup = bs(html, "html.parser")
        title = soup.find("h1", {"ArticlePage-headline"}).get_text()
        title = title.strip().replace('\xa0','')
        content_total.extend([ p.get_text() for p in soup.find_all('p')])
        content_total = content_total[0:-7]
        content = ' '.join(content_total)
        content = content.replace('\xa0','')
        if n%1000 == 0:
            df7 = pd.DataFrame(res_list)
            df7.to_excel('latimes(7).xlsx', sheet_name = 'sheet1')
    except Exception as e:
        print(e)
        continue
    res_list.append({'title' : title, 'content': content})
    print('{}번 페이지 완료'.format(n))
```

```python
df6 = pd.DataFrame(res_list)
df6.to_excel('latimes(6).xlsx', sheet_name = 'sheet1')
```
