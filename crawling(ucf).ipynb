```python
# _*_ coding: UTF-8 _*_
#coding:utf-8
from urllib.request import urlopen
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd
import csv
from selenium.webdriver.remote.webelement import WebElement
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import re
import numpy as np
```

```python
#매크로용 웹켜기
#driver = webdriver.Remote('http://localhost:4444/wd/hub', options.to_capabilities())  # 셀레 서버를 Standalone 방식으로 가동
driver = webdriver.Chrome('./chromedriver')
driver.implicitly_wait(5)
```

```python
#원하는 URL 열기
driver.get('https://knightconnect.campuslabs.com/engage/organizations')
```

```python
soup = bs(driver.page_source, 'html.parser')
```

```python
soup.find("a", {"name":"display: block; text-decoration: none; margin-bottom: 20px;"}) # 메타는 태그 이름일 뿐, 헤드에서 메타 태그 다 가져오기!
```

```python
soup
```

```python
for href in soup.find_all('a'): ##속성에 있는 값 가져오기 in 이하에 있는 애들 안에 meta 중 첫번째
    print(href.get('href'))
```

```python
href_list = []   #print(href.get('href'))
href_list.extend([ i.get('href') for i in soup.find_all('a')])
```

```python
href_list1 = href_list[10:-3]
```

```python
href_list1
```

```python
res_list = []
for article in href_list1:
        driver.get('https://knightconnect.campuslabs.com' + article)
        soup1 = bs(driver.page_source, 'html.parser')
        try:
            title = soup1.find("h1", {"style":"padding: 13px 0px 0px 85px;"}).get_text()
            soup2 = soup1.find("div", {"style":"margin-left: 5px; padding: 5px 15px; border-left: 1px solid rgb(210, 210, 210);"})
            email = soup2.select('div')[-1].get_text()[19:-1]
        except Exception as e:
            print(e)
            continue
        res_list.append({'title' : title, 'email' : email})
```
